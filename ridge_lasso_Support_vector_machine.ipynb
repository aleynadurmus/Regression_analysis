```
import mglearn
```

```
from sklearn.model_selection import train_test_split
x,y=mglearn.datasets.load_extended_boston()
```

```
x_egitim,x_test,y_egitim,y_test=train_test_split(x,y,random_state=0)
```

```
from sklearn.linear_model import Ridge
```

```
ridge=Ridge().fit(x_egitim,y_egitim)
```

```
print(ridge.score(x_egitim,y_egitim))
print(ridge.score(x_test,y_test))
```

```
from sklearn.linear_model import Lasso
```

```
lasso=Lasso().fit(x_egitim,y_egitim)
```

```
print(lasso.score(x_egitim,y_egitim))
print(lasso.score(x_test,y_test))
```

```
import numpy as np
```

```
np.sum(lasso.coef_!=0)
```

```
lasso0=Lasso(alpha=0.001,max_iter=100000).fit(x_egitim,y_egitim)
print(lasso0.score(x_egitim,y_egitim))
print(lasso0.score(x_test,y_test))
```

```
from sklearn.datasets import load_breast_cancer
```

```
kanser=load_breast_cancer()
```

```
from sklearn.model_selection import train_test_split
```

```
x_egitim,x_test,y_egitim,y_test=train_test_split(kanser.data,kanser.target, stratify=kanser.target,random_state=42)
```

```
from sklearn.linear_model import LogisticRegression
```

```
logreg=LogisticRegression(C=100,solver='liblinear').fit(x_egitim,y_egitim)
```

```
print(logreg.score(x_egitim,y_egitim))
print(logreg.score(x_test,y_test))
```
