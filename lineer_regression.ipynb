```
import mglearn
from sklearn.linear_model import LinearRegression
```

```
x,y=mglearn.datasets.make_wave(n_samples=60)
```

```
from sklearn.model_selection import train_test_split
```

```
x_egitim,x_test,y_egitim,y_test=train_test_split(x,y,random_state=40)
```

```
lr=LinearRegression().fit(x_egitim,y_egitim)
```

```
print(lr.coef_)
```

```
print(lr.intercept_)
```

```
print(lr.score(x_egitim,y_egitim))
```

```
print(lr.score(x_test,y_test))
```

```
import pandas as pd
```

```
veri=pd.read_csv('/content/drive/MyDrive/calisma_kodlari/student-mat.csv',sep=';')
```

```
veri=veri[['G1','G2','G3','studytime','failures','absences','age']]
```

```
veri.head()
```

```
import numpy as np
y=np.array(veri['G3'])
x=np.array(veri.drop('G3',axis=1))
```

```
from sklearn.model_selection import train_test_split
```

```
x_egitim,x_test,y_egitim,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```
linear=LinearRegression()
```

```
linear.fit(x_egitim,y_egitim)
```

```
linear.score(x_test,y_test)
```

```
linear.score(x_egitim,y_egitim)
```

```
print(linear.coef_)
```

```
print(linear.intercept_)
```

```
yeni=np.array([[71,54,5,0,0,18]])
linear.predict(yeni)
```
